Carregar a base TITANIC do seaborn

```python
# Para carregar um dataset do Seaborn usar
from seaborn import load_dataset
#import pandas as pd

# Para usar um dataset do Seaborn criar um Dataframe(df)
df = load_dataset('titanic')

#df = pd.DataFrame(data = df)

df
```

Observar a relação dos atributos "pclass" com "class".
Foi observadi que os valores são correspondente, neste caso a solução foi suprimir a coluna "class"

```python
# Eliminar a coluna 'class' que tem os mesmos valores de 'pclass', 'axis=1' indica que vai apagar a coluna
# 'axis=0' navegação horizontal(Linhas), 'axis=1' navegação vertical(Colunas)
df = df.drop('class', axis = 1)

df
```

Observar a quantidade de NaN existente no atributo "deck"; Devido a falta de dados ou a mioria com o valor 'NaN', a coluna 'deck' também vai ser removida.

Remover as linhas que teham valores NaN

```python
# 'dropna' remove as linhas que contem NaN como valores
df = df.dropna()

df
```

Transformar todos os atributos categóricos usando One-Hot Encoder; Foram identificado as colunas 'sex1', 'embarked', 'who', 'adult_male', 'embark_town', 'alive' e 'alone' com valores categóricos, assim estou dividindo em 2 grupos, grupo 1 (Colunas com apenas 2 valores categóricos, como é o caso 'sex', 'adult_male', 'alive' e 'alone'), e grupo 2('embarked', 'who', 'embark_town'), assim pretendo usar Label Encoder para o gurpo 1 e One Hot Encoder para o grupo 2.

```python
from sklearn import preprocessing
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

df = pd.DataFrame(data = df)

# Groupo 1 com uso do Label Encoder
label_encoder = preprocessing.LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['adult_male'] = label_encoder.fit_transform(df['adult_male'])
df['alive'] = label_encoder.fit_transform(df['alive'])
df['alone'] = label_encoder.fit_transform(df['alone'])

# Groupo 2 com uso do One Hot Encoder
one_hot_encoder = make_column_transformer((OneHotEncoder(), ['embarked','who', 'embark_town']), remainder='passthrough')
#one_hot_encoder = pd.get_dummies(df, columns = ['embarked','who', 'embark_town'])
# Colunas 0, 1, 2 = 'embarked', colunas 3, 4, 5 = 'who', colunas 6, 7, 8 = 'embark_town
# Coluna 9 = 'survived', coluna 10 = 'pclass'
# Coluna 11 = 'sex', coluna 12 = 'age', coluna 13 = 'sibsp', coluna 14 = 'parch'
# Coluna 15 = 'fare', coluna 16 = 'adult_male', coluna 17 = 'alive', coluna 18 = 'alone'

df = one_hot_encoder.fit_transform(df)
df = pd.DataFrame(data = df)
df
```

Coluna target é a "pclass" ou "survived"; a coluna "pclass" = coluna 10 e coluna "survived" = coluna 9, a escolha para a target vai ser a coluna 10.

```python
y = df[10]

X = df.drop(10, axis=1)

y
```

Separar os dados de treinamento e testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = None, stratify = y)
```

Treinar 2 árvores de decisão com características diferentes

Treinar 2 kNN com características diferentes

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Árvore de Decisão
treino1 = DecisionTreeClassifier(criterion = "entropy")
treino2 = DecisionTreeClassifier(max_depth = 3)

# Protocolo KNN
treino3 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean', algorithm = 'brute')
treino4 = KNeighborsClassifier(n_neighbors = 7, metric = 'manhattan', algorithm = 'brute')

treino1.fit(X_train, y_train)
treino2.fit(X_train, y_train)
treino3.fit(X_train, y_train)
treino4.fit(X_train, y_train)
```

```python
from sklearn import metrics

result1 = treino1.predict(X_test)
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Árvore Decisão Profundidade livre: " + "{}%".format(show1))

result2 = treino2.predict(X_test)
acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Árvore Decisão Profundidade 3: " + "{}%".format(show1))

result3 = treino3.predict(X_test)
acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Protocolo KNN Metric 'eucledean': " + "{}%".format(show3))

result4 = treino4.predict(X_test)
acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Protocolo KNN Metric 'manhattan': " + "{}%".format(show3))
```
